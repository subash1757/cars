```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
Data = pd.read_csv("Cars.csv")
```

## Count vehicle produced  for each year

```python
Data["year"].value_counts()
```

## Count of vehicle produced for each year in bar plot

```python
Data["year"].value_counts().plot(kind="bar")
```

## Count of fuel type

```python
Data["fuel"].value_counts().plot(kind="bar")
```

## Diesel engine car prize maximum and minimum

```python
Deisel_list = Data[Data['fuel']=='Diesel']
max_price_Diesel = Deisel_list["selling_price"].max()
name = Deisel_list[Deisel_list["selling_price"]==max_price_Diesel]
maximum_price_Diesel = name[["name","selling_price"]]

maximum_price_Diesel.columns = ["Veichle name","Maximum selling price"]
maximum_price_Diesel
```

```python
Deisel_list = Data[Data['fuel']=='Diesel']
min_price_Diesel = Deisel_list["selling_price"].min()
name = Deisel_list[Deisel_list["selling_price"]==min_price_Diesel]
minimum_price_Diesel = name[["name","selling_price"]]

minimum_price_Diesel.columns = ["Veichle name","Minimum selling price"]
minimum_price_Diesel
```

## Pei-chart to show the seller_type

## Count of transmission 

```python
Data["transmission"].value_counts().plot(kind="bar")
```

## Finding a distribution between owners and selling price 

```python
plt.figure(figsize=(15,15))
sns.boxplot(x="owner",y="selling_price",data=Data);
```

## Counting transmission based on the years

```python
plt.figure(figsize=(15,20))
sns.countplot(x="year",hue="transmission",data=Data);
plt.ylabel("transmission");
```

## Ploting year distribution

```python
sns.distplot(x=Data["year"],bins=20);
```

## summarized whole dataset

```python
plt.figure(figsize=(15,15))
sns.pairplot(Data, hue='fuel', size=2.5);
```
